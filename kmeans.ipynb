```python
import pandas as pd
import numpy as np
data=pd.read_csv('income.csv')
data.head(10)
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['Age'],data['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income($)')
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3)
km.fit_predict(data[['Age']],data[['Income($)']])
```

```python
y_predicted=km.fit_predict(data[['Age']],data[['Income($)']])
```

```python
data['cluster']=y_predicted
```

```python
data
```

```python
data1=data[data.cluster==0]
data2=data[data.cluster==1]
data3=data[data.cluster==2]
```

```python
%matplotlib inline
plt.scatter(data1['Age'],data1['Income($)'],color="red",label="cluster 1")
plt.scatter(data2['Age'],data2['Income($)'],color="blue",label="cluster 2")
plt.scatter(data3['Age'],data3['Income($)'],color="green",label="cluster 3")
plt.xlabel('Age')
plt.ylabel('Income($)')
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(data[['Age']])
data['Age']=scaler.transform(data[['Age']])
scaler.fit(data[['Income($)']])
data['Income($)']=scaler.transform(data[['Income($)']])
data
```

```python
data=data.drop(['cluster'],axis="columns")
```

```python
data
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3)
y_predicted=km.fit_predict(data[['Age','Income($)']])
```

```python
data['cluster']=y_predicted
```

```python
data1=data[data.cluster==0]
data2=data[data.cluster==1]
data3=data[data.cluster==2]
```

```python
km.cluster_centers_
```

```python
%matplotlib inline
plt.scatter(data1['Age'], data1['Income($)'], color="red", label="cluster 1")
plt.scatter(data2['Age'], data2['Income($)'], color="blue", label="cluster 2")
plt.scatter(data3['Age'], data3['Income($)'], color="green", label="cluster 3")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color="purple",marker="x", label="centeroid")
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
sse = []
k_rng=range(1,10)
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(data[['Age']],data[['Income($)']])
    sse.append(km.inertia_)
```

```python
plt.plot(k_rng,sse)
plt.xlabel('k')
plt.ylabel('sse')
```
